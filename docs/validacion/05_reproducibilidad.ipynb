# 05. Reproducibilidad por semilla

Valida que dos ejecuciones con la misma semilla producen resultados **bit a bit idénticos**.

Mesa 3.x tiene un detalle conocido: `seed=` no controla el `random` global de Python. El modelo lo arregla llamando explícitamente a `random.seed()` y `np.random.seed()` en su constructor (ver `src/model.py`).


```python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parents[1]))

import pandas as pd
from src.model import MarketModel

def run(seed=42, days=200):
    m = MarketModel(seed=seed)
    for _ in range(days):
        m.step_day()
    return m.datacollector.get_model_vars_dataframe()

df1 = run()
df2 = run()
print("Idénticos:", df1.equals(df2))
```
